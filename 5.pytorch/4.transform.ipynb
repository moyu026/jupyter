## 1.裁剪

### 1.1 transforms.Resize
- transforms.Resize(256): 按图片宽高比将图片短边缩放到256
- transforms.Resize((256,256)): 按图片宽和高同时缩放到256

```python
# transforms.Resize(256):按图片宽高比将图片短边缩放到256
import matplotlib.pyplot as plt
from PIL import Image
from torchvision import transforms

file_path = "./file/1.jpg"
img = Image.open(file_path)
print("origin_img_size:", img.size)

trans = transforms.Compose([transforms.Resize(256),
                            ])
img1 = trans(img)
print("变换后的大小:", img1.size)
plt.subplot(1, 2, 1), plt.imshow(img)
plt.subplot(1, 2, 2), plt.imshow(img1)
plt.show()
```

```python
# transforms.Resize((256,256)):按图片宽和高同时缩放到256
import matplotlib.pyplot as plt
from PIL import Image
from torchvision import transforms

file_path = "./file/1.jpg"
img = Image.open(file_path)
print("origin_img_size:", img.size)

trans = transforms.Compose([transforms.Resize((256,256)),
                            ])
img1 = trans(img)
print("变换后的大小:", img1.size)
plt.subplot(1, 2, 1), plt.imshow(img)
plt.subplot(1, 2, 2), plt.imshow(img1)
plt.show()
```

### 1.2 transforms.RandomResizedCrop
-  将原图片随机裁剪出一块,位置随机，大小也随机，再缩放成相应 (size*size) 的比例

```python
# transforms.Resize((256,256)):按图片宽和高同时缩放到256
import matplotlib.pyplot as plt
from PIL import Image
from torchvision import transforms

file_path = "./file/1.jpg"
img = Image.open(file_path)
print("origin_img_size:", img.size)

trans = transforms.Compose([transforms.RandomResizedCrop(256),
                            ])
img1 = trans(img)
print("变换后的大小:", img1.size)
plt.subplot(1, 2, 1), plt.imshow(img)
plt.subplot(1, 2, 2), plt.imshow(img1)
plt.show()
```

### 1.3 transforms.CenterCrop
- 从图片中心开始沿两边裁剪，裁剪后的图片大小为（size*size）
- 如果图片宽或高比裁剪的size小，则补黑边

```python
import matplotlib.pyplot as plt
from PIL import Image
from torchvision import transforms

file_path = "./file/1.jpg"
img = Image.open(file_path)
print("origin_img_size:", img.size)

trans = transforms.Compose([transforms.CenterCrop(300)
                           ])
img1 = trans(img)
print("变换后的大小:", img1.size)
plt.subplot(1, 2, 1), plt.imshow(img)
plt.subplot(1, 2, 2), plt.imshow(img1)
plt.show()
```

### 1.4 transforms.RandomCrop
- 对图像随机裁剪出尺寸为size的图片，裁剪位置随机，裁剪尺寸只能小于高和宽

```python
import matplotlib.pyplot as plt
from PIL import Image
from torchvision import transforms

file_path = "./file/1.jpg"
img = Image.open(file_path)
print("origin_img_size:", img.size)

trans = transforms.Compose([transforms.RandomCrop(100)
                           ])
img1 = trans(img)
print("变换后的大小:", img1.size)
plt.subplot(1, 2, 1), plt.imshow(img)
plt.subplot(1, 2, 2), plt.imshow(img1)
plt.show()
```

## 2.旋转
- 1.transforms.RandomHorizontalFlip
- 2.transforms.RandomVerticalFlip
- 3.transforms.RandomRotation

### 2.1 transforms.RandomHorizontalFlip
- 根据概率p对图片进行水平（左右）翻转，每次根据概率来决定是否执行翻转

```python
import matplotlib.pyplot as plt
from PIL import Image
from torchvision import transforms

file_path = "./file/1.jpg"
img = Image.open(file_path)
print("origin_img_size:", img.size)

trans = transforms.Compose([transforms.RandomHorizontalFlip(p=1)
                           ])
img1 = trans(img)
print("变换后的大小:", img1.size)
plt.subplot(1, 2, 1), plt.imshow(img)
plt.subplot(1, 2, 2), plt.imshow(img1)
plt.show()
```

### 2.2 transforms.RandomVerticalFlip
- 根据概率p对图片进行垂直（上下）翻转，每次根据概率来决定是否执行翻转

```python
import matplotlib.pyplot as plt
from PIL import Image
from torchvision import transforms

file_path = "./file/1.jpg"
img = Image.open(file_path)
print("origin_img_size:", img.size)

trans = transforms.Compose([transforms.RandomVerticalFlip(p=1)
                           ])
img1 = trans(img)
print("变换后的大小:", img1.size)
plt.subplot(1, 2, 1), plt.imshow(img)
plt.subplot(1, 2, 2), plt.imshow(img1)
plt.show()
```

### 2.3 transforms.RandomRotation
- 对图片旋转随机的角度degrees
- I. 当degrees为a时，在区间(-a,a)之间随机选择旋转角度；
- II. 当degrees为(a,b)时，在区间(a,b)之间随机选择旋转角度；

```python
import matplotlib.pyplot as plt
from PIL import Image
from torchvision import transforms

file_path = "./file/1.jpg"
img = Image.open(file_path)
print("origin_img_size:", img.size)

trans = transforms.Compose([transforms.RandomRotation(degrees=50)
                           ])
img1 = trans(img)
print("变换后的大小:", img1.size)
plt.subplot(1, 2, 1), plt.imshow(img)
plt.subplot(1, 2, 2), plt.imshow(img1)
plt.show()
```
