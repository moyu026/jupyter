```python
import cv2
from skimage import morphology
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_mask = cv2.imread('./image_file/mask/pointer_msk.jpg',0)  # 0表示以灰度图的方式读取图片
img_mask = cv2.resize(img_mask,(512,512))
# img_mask = cv2.cvtColor(img_mask, cv2.COLOR_GRAY2BGR)
# img_mask.shape
plt.imshow(img_mask, cmap='gray')
```

### 骨架算法

```python
pointer_skeleton = morphology.skeletonize(img_mask)
# plt.imshow(pointer_skeleton, cmap='gray')
# pointer_skeleton.shape
pointer_skeleton
```

```python
pointer_edges = pointer_skeleton * 255
pointer_edges = pointer_edges.astype(np.uint8)
plt.imshow(pointer_edges, cmap='gray')
```

### lines = cv2.HoughLinesP(image, rho, theta, threshold, minLineLength, maxLineGap)
>image：输入的二值化图像（一般为边缘检测后的图像，如Canny边缘检测后的结果）。
rho：距离参数 $\rho$ 的精度，一般取1。
theta：角度参数 $\theta$ 的精度，一般取$\pi/180$。
threshold：用于确定直线的最小投票数。
minLineLength：线的最小长度，小于此长度的线段将被忽略。
maxLineGap：同一条线上两点之间的最大允许间隔，超过此值将被认为是不同的线段。

```python
pointer_lines = cv2.HoughLinesP(pointer_edges, 1, np.pi / 180, 10, np.array([]), minLineLength=150,maxLineGap=400)
pointer_lines
```

```python
for x1, y1, x2, y2 in pointer_lines[0]:
    coin1 = (x1, y1)
    coin2 = (x2, y2)
    cv2.line(img_mask, (x1, y1), (x2, y2), (0, 0, 255), 2)
plt.imshow(img_mask, cmap='gray')
```
