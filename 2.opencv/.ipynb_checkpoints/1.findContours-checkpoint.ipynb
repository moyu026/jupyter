```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_path = r'./image_file/findContours/fire.png'
img = cv2.imread(img_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # 将BGR格式转换成灰度图片
plt.imshow(gray, cmap='gray')
```

### 二值化，二值化的输入图像是灰度图

```python
ret, binary = cv2.threshold(gray, 10, 255, cv2.THRESH_BINARY) # ret:设定的阈值， binary：二值化后的图像
```

### findContours的输入可以是灰度图或者二值化后的图
### 用drawContours函数画出检测出的轮廓

```python
kernel = np.ones((1, 5), np.uint8)
binary = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel, anchor=(2, 0), iterations=5)
contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE) 
cv2.drawContours(img, contours, -1, (0, 0, 255), 3)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

### 计算选中轮廓的面积的大小

```python
for i in contours:
    area = cv2.contourArea(i)
    print(area)
```
